```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Create subplots
fig, axs = plt.subplots(2, 1)

# Line plot
axs[0].plot(x, y)
axs[0].set_title("Line Plot")

# Scatter plot
axs[1].scatter(x, y)
axs[1].set_title("Scatter Plot")

# Display plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
months = [
    "2025-01", "2025-02", "2025-03", "2025-04", "2025-05", "2025-06",
    "2025-07", "2025-08", "2025-09", "2025-10", "2025-11", "2025-12"
]
sales = [
    67850.30, 63240.75, 71450.90, 76890.25, 43210.50, 82345.60,
    89760.35, 86430.20, 82150.75, 91240.30, 103560.45, 118790.80
]

# Create a figure and plot
plt.figure(figsize=(12,6))

# Plotting the data with customized styling
plt.plot(months, sales, color="purple", markersize=10, marker="o", linestyle="-", linewidth=2)

# Title and labels
plt.title("Sales of Product in 2025", fontsize=16, fontweight='bold')
plt.xlabel("Months", fontsize=14)
plt.ylabel("Sales (in USD)", fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adding a grid for better visualization
plt.grid(True)

# Display plot
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
de_nepal = [10.5, 25.21, 58.65, 119.27, 274.87]

# Create the plot
plt.figure(figsize=(10,6))

# Plot India and Nepal energy consumption
plt.plot(year, e_india, label='India', color='blue', marker='o', linestyle='-', linewidth=2, markersize=8)
plt.plot(year, de_nepal, label='Nepal', color='green', marker='s', linestyle='-', linewidth=2, markersize=8)

# Adding title and labels
plt.title("Energy Consumption Over the Years", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Energy Consumption (in TWh)", fontsize=14)

# Add a legend
plt.legend()

# Add a grid for better visualization
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
